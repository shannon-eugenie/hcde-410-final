##### **Cleaning Data**
So, to start analyzing the dataset, the code below works to load the Heart Health dataset CSV file and turn it into a data frame for easier handling.

```python
import pandas as pd 
def load_data_bmi(csv_path, delimiter):
    return pd.read_csv(csv_path, delimiter=delimiter)

load_data_bmi("Heart_health.csv", ",")
```

To check for duplicates within the dataset, *duplicated()* will be used. This function checks for any duplicated rows. Then the code will print out the duplicated rows found.

```python
heart_data = load_data_bmi("Heart_health.csv", ",")

# Counts the total number of duplicate rows
print(heart_data.duplicated().sum())

# Prints the duplicated rows
print(heart_data[heart_data.duplicated(keep=False)])
```

There seem to be 10 pairs of duplicated data (20 rows in total) found in the dataset. So, *drop_duplicates()* can be used to get rid of all the duplicated data, and only keep the unique rows/data. To make sure that the duplicates are truly gone, *duplicated()* can be used again.

```python
clean_heart_data = load_data_bmi("Heart_health.csv", ",").drop_duplicates()

print(clean_heart_data.duplicated().sum())
```

##### **Processing Data**
Now that the data has been cleaned, this code functions to create a new column calculating the body mass index (BMI) from the *'Height(cm)'* and *'Weight(kg)'* fields using the formula: *$weight(kg)/height(m)^2$*. To further process the data for analysis, the *'Smoker'* field will be converted from yes/no values (strings) to binary values. Yes will become 1 (smoker) and No will become 0 (non-smoker). Additionally, the *'Blood Pressure(mmHg)'* field values will also be separated into 2 new fields *'Systolic Pressure(mmHg)'* and *'Diastolic Pressure(mmHg)'*. *'Systolic Pressure(mmHg)'* represents the first number that measures the pressure when the heart beats and *'Diastolic Pressure(mmHg)'* represents the second number that measures the pressure when the heart rests between beats [4]. Then, the code will delete the *'Blood Pressure(mmHg)'*, *'Height(cm)'*, and *'Weight(kg)'* fields, along with the newly made/converted *'Height(m)'* field. 

*ChatGPT is used to process the 'Smoker' field and separate the 'Blood Pressure(mmHg)' field into 2 new separate fields*

```python
# Convert Height data from cm to m
clean_heart_data['Height(m)'] = clean_heart_data['Height(cm)'] / 100

# Calculate BMI and round to 1 decimal place
clean_heart_data['BMI'] = (clean_heart_data['Weight(kg)'] / (clean_heart_data['Height(m)'] ** 2)).round(1)

# Converts Smoker yes/no values to binary values
clean_heart_data['Smoker'] = clean_heart_data['Smoker'].map({'Yes': 1, 'No': 0})

# Splits the Blood Pressure field into 2 new fields
clean_heart_data[['Systolic Pressure(mmHg)', 'Diastolic Pressure(mmHg)']] = clean_heart_data['Blood Pressure(mmHg)'].str.split('/', expand=True)

# Deletes the Blood Pressure, Height, and Weight field
clean_heart_data = clean_heart_data.drop(['Blood Pressure(mmHg)', 'Height(m)', 'Height(cm)', 'Weight(kg)'], axis=1)

print(clean_heart_data)
```

Since research question 1 focuses on identifying the age and gender distribution of individuals with a risk of heart attack, the dataset is filtered out to only take in the data where heart attack = 1 (at risk). From this newly processed data, the gender and age distribution of individuals with a risk of heart attack can then be identified.

```python
# Filters data for individuals at risk of heart attack
heart_attack = clean_heart_data[clean_heart_data['Heart Attack'] == 1]

print(heart_attack)
```

## Findings
### Q1: What is the age and gender distribution of individuals with a risk of heart attack?
#### Age Distribution
This code creates a histogram displaying the age distribution for individuals with risk of heart attack. It will also print out the exact counts for each age.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Generates a histogram
sns.set_style('whitegrid')
plt.figure(figsize=(10,6))

sns.histplot(heart_attack['Age'], kde=True, bins=30)

plt.title("Age Distribution of Individuals with a Risk of Heart Attack")
plt.xlabel("Age")
plt.ylabel("Count")

plt.show()

print(heart_attack['Age'].value_counts())
```

The age distribution for individuals at risk of a heart attack seems to be skewed to the late 40s and mid-50s, peaking at the age of 47. When calculated individually (amount per age), the risk of heart attack peaks at the age of 47 with a total of 59 individuals and at age 55 with 50 individuals.

#### Gender Distribution
The code below functions to create a bar plot showing the distribution between genders for individuals with a risk of heart attack and prints out the exact count for males and females.

```python
# Generates a bar plot
sns.set_style('whitegrid')

plt.figure(figsize=(10,6))
sns.countplot(data=heart_attack, x='Gender')
plt.title("Gender Distribution of Individuals with a Risk of Heart Attack")
plt.xlabel("Gender")
plt.ylabel("Count")

plt.show()

# Prints the actual amount for each gender
print(heart_attack['Gender'].value_counts())
```

### Q2: What is the correlation/relationship between each lifestyle factor (e.g., smoking, exercise) and physical health indicator (e.g., cholesterol levels, blood pressure (systolic and diastolic pressure), glucose levels, height and weight (BMI)) with the risk of heart attack?

To answer this question, *clean_heart_data* will be used to include both individuals with and without the risk of heart attack. The code below functions to loop through all the fields one by one creating a line graph for each factor/indicator as the x-axis and risk of heart attack as the y-axis. It will also check through the fields one by one, specifically looking for the fields *'Systolic Pressure(mmHg)'* and *'Diastolic Pressure(mmHg)'* to sort out the data in ascending order when displayed on the graph.

*ChatGPT is used to help create this graph*

```python
import numpy as np

variables = ['Cholesterol(mg/dL)', 'Glucose(mg/dL)', 'Smoker', 'Exercise(hours/week)', 
                   'BMI', 'Systolic Pressure(mmHg)', 'Diastolic Pressure(mmHg)']

fig, axes = plt.subplots(len(variables), 1, figsize=(10, 18))

# Loop to create a line graph for each factor/indicator vs heart attack
for idx, var in enumerate(variables):
    ax = axes[idx]
    
    # Sorts data in ascending order for systolic and diastolic pressure fields
    if var in ['Systolic Pressure(mmHg)', 'Diastolic Pressure(mmHg)']:
        sorted_data = clean_heart_data.sort_values(by=var)  
    else:
        sorted_data = clean_heart_data

    # Create the line graph
    sns.lineplot(data=sorted_data, x=var, y='Heart Attack', ax=ax)
    ax.grid(True)
    
    ax.set_title(f"{var} vs Heart Attack")
    ax.set_xlabel(var)
    ax.set_ylabel("Heart Attack Risk (0 or 1)")

plt.tight_layout()
plt.show()
```

There seems to be a positive correlation between all the factors/indicators and the risk of heart attack.

1. **Cholesterol(mg/dL) vs Heart Attack**
  
   The graph indicates that cholesterol levels above 200 mg/dL are associated with an increased risk of heart attack as there is a sharp increase, stabilizing at a high risk beyond 210 mg/dL. Levels of 195 mg/dL or lower show no significant risk.
   
2. **Glucose(mg/dL) vs Heart Attack**

   The glucose level starts to spike significantly from 90 to around 94 mg/dL. This suggests that glucose levels exceeding 95 mg/dL are linked to a higher risk, while levels of 88 mg/dL or less show minimal risk.

3. **Smoker vs Heart Attack**

   There seems to be a slight increase in heart attack risk as the smoking status moves from non-smoker (0) to smoker (1). However, this graph has the most range of uncertainty compared to the other factors/indicators, as shown by the shaded lighter blue section that represents the confidence interval. Despite the uncertainty and variability, the general positive increase trend remains consistent. 
      
4. **Exercise(hours/week) vs Heart Attack**

   Counterintuitively, there seems to be an upward increase in heart attack risk with higher hours of exercise per week. The graph indicates that individuals exercising more hours per week (4 hours/week) show higher heart attack risk compared to those exercising fewer hours (1 hour/week).
   
5. **BMI vs Heart Attack**

   The graph shows a lot of heart attack risk fluctuation, spiking randomly between BMIs of 24.5 and 26.4 before stabilizing at consistently high risk above 26.4.
   
6. **Systolic Pressure(mmHg) vs Heart Attack**

    The risk of heart attack starts to increase at the systolic pressure of 120 mmHg, slightly decreases at 12 mmHg, and spikes up again at 122 mmHg. Higher systolic pressure seems to lead to a higher risk of heart attack as it presents a consistent 1 for the risk of heart attack (1 = risk of heart attack is present) at 123 mmHg.
   
7. **Diastolic Pressure(mmHg) vs Heart Attack**

   From the graph, the risk of heart attack escalates at the diastolic pressure of 76 mmHg, drops slightly at 77 mmHg, and increases at 78 mmHg. This indicates that the higher the diastolic pressure, the higher the risk of heart attack as it stabilizes at 79 mmHg (all results in 1 for heart attack, meaning the risk is present).

#### Q3: What is the main contributor to the increased risk of heart attack?

To find the main contributor and further expand on the correlation/relationship findings between each factor/indicator vs. the risk of heart attack, *clean_heart_data* will be used once again to include the data for both individuals with and without risk of heart attack. The code below functions to calculate the correlation matrix between each factor/indicator to the risk of heart attack.

```python
# Calculate the correlation matrix
correlation_matrix = clean_heart_data[['Cholesterol(mg/dL)', 'Glucose(mg/dL)', 'Smoker', 
                                 'Exercise(hours/week)', 'BMI', 'Systolic Pressure(mmHg)', 
                                 'Diastolic Pressure(mmHg)', 'Heart Attack']].corr()

print(correlation_matrix['Heart Attack'])
```

The correlation matrix indicates that the factor/indicator with the highest correlation (rounded up) to the increased risk of heart attack is cholesterol levels at 0.83, followed by glucose levels at 0.79, diaslotic pressure at 0.76, and systolic pressure at 0.72. Smoking demonstrates the weakest correlation at 0.02, indicating minimal influence in this dataset. This suggests that the main contributor to a higher risk of heart attack may be cholesterol level, followed by glucose level and blood pressure (systolic and diastolic pressure).

## Discussion *(Limitations and Implications)*
#### Q1: What is the age and gender distribution of individuals with a risk of heart attack?
The findings indicate that the majority of individuals at risk of heart attack are in their late 40s and mid-50s, peaking at age 47. This aligns with existing research [2][3], which suggests that heart attacks predominantly affect individuals aged 45 and older, likely due to the cumulative effects of lifestyle factors and age-related physiological changes. Furthermore, the results show a male-dominant heart attack risk distribution, with significantly more men than women identified at risk. This may reflect biological differences, such as variations in hormonal protection or differences in cardiovascular risk profiles between genders. According to [5], "Men are at increased risk of heart attack earlier in life due to hormone changes, abdominal obesity, and emotional challenges in middle adulthood."

However, this dataset may have an uneven distribution of participants across age and gender, which may introduce bias. For example, an underrepresentation of women in certain age brackets might skew the gender-related findings, exaggerating the male-dominant heart attack risk distribution; this could explain why the gender distribution is very uneven. Thus, improvements can be made by analyzing the age and gender distribution in a single graph to fully understand the combined effect of these variables on heart attack risk. Additionally, the sample size may not accurately capture the diversity within the population, limiting the generalizability of the results. These findings highlight the importance of applying targeted preventative strategies, especially for middle-aged individuals. It pushes the importance of routine screenings for individuals in their 40s and 50s for preventative measures. Future studies should ensure balanced demographic representation to provide a more accurate picture of age and gender-based distributions. 

#### Q2: What is the correlation/relationship between each lifestyle factor (e.g., smoking, exercise) and physical health indicator (e.g., cholesterol levels, blood pressure (systolic and diastolic pressure), glucose levels, height and weight (BMI)) with the risk of heart attack?
All analyzed factors/indicators exhibit a positive correlation with heart attack risk, albeit to varying degrees. Cholesterol levels above 200 mg/dL are strongly associated with increased heart attack risk, with consistent risk at levels above 210 mg/dL. This suggests that cholesterol management is crucial in mitigating cardiovascular risks. Glucose levels above 95 mg/dL are linked to elevated heart attack risk, while lower levels (under 88 mg/dL) show minimal risk. This reinforces the role of glucose in conditions such as diabetes that may contribute to the development of cardiovascular diseases. Smoking shows a positive correlation, although there's only a slight increase. BMI fluctuations between 24.5 and 26.4 indicate instability in heart attack risk before fully stabilizing at a consistently higher risk above 26.4. This suggests that obesity may contribute to increased cardiovascular risk. Elevated blood pressure also strongly correlates with heart attack risk. Systolic pressure beyond 120 mmHg and diastolic pressure beyond 76 mmHg show significant heart attack risk, consistent with hypertension/high blood pressure as a critical cardiovascular risk factor [3]. Unexpectedly, the analysis reveals a positive correlation between exercise and heart attack risk, suggesting that individuals who exercise more frequently (4 hours/week) have a higher risk. This counterintuitive finding may stem from confounding factors such as the intensity of exercise, pre-existing health conditions, or the possibility of overexertion. High-intensity workouts can strain the cardiovascular system (increasing the heart rate and blood pressure), potentially triggering issues in at-risk individuals. 

The data might have sample/selection bias where there is a higher prevalence of individuals with health issues that predispose them to heart attacks, regardless of exercise, thus causing an inaccurate interpretation/finding for exercise vs. risk of heart attack. According to [3], a lack of physical activity is said to raise the risk of a heart attack as it makes it more likely to develop coronary artery disease, which is the opposite of the finding found in this analysis. Overall, cholesterol, glucose levels, and blood pressure (systolic and diastolic pressure) should remain a focus for preventive care for heart attacks. 

#### Q3: What is the main contributor to the increased risk of heart attack?
The correlation matrix calculation reveals that cholesterol levels have the highest correlation with heart attack risk (0.83), followed by glucose levels (0.79), diastolic pressure (0.76), and systolic pressure (0.72). Smoking shows the weakest correlation among all factors/indicators (0.02), suggesting a minimal impact in this dataset. These results indicate that cholesterol levels are likely the primary contributor to increased heart attack risk. Elevated cholesterol levels can lead to plaque buildup in arteries, a major cause of cardiovascular diseases. Similarly, high glucose levels and hypertension/high blood pressure (both systolic and diastolic pressure) contribute significantly to cardiovascular strain [3]. 

The main limitation of this analysis is the reliance on linear correlations, which may oversimplify and fail to capture the complex, multifactorial relationships between risk factors/indicators. Heart attack risk is likely influenced by numerous interdependent factors, such as the combined effect of high cholesterol and glucose levels. Linear correlations fail to account for these interactions and may underestimate the contribution of factors that have indirect or conditional effects on heart attack risk. Other potential contributors, such as genetic predisposition, psychosocial factors (stress), etc., are also not being studied in this analysis. So, the conclusion for the main contributor may not be as accurate. In addition, the study population may not be representative of broader demographics, thus limiting external validity. Therefore, targeting the management of cholesterol, glucose levels, and blood pressure should be central to cardiovascular health programs. Individuals should be educated about these critical indicators to better promote early detection and intervention. Additionally, the low correlation of smoking emphasizes the need to revisit the dataset to check for any potential biases or confounding factors. However, an important thing to note is that correlation does not imply causation, so it cannot truly be concluded that any single factor/indicator directly causes an increased risk of heart attack.
